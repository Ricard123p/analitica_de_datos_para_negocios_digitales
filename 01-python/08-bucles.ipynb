### Bucle infinito

````python
while True:
   print('Estoy dentro del bucle y no puedo salir')
````

````python
numero_grande = -999999999

# ingresa el primer valor 
number = int(input('Introduce un numero o escribe -1 para terminar'))

while number!= -1:
    # Es el numero mas que el valor numerico?
    if number > numero_grande:
        numero_grande = number
    number = int(input('Introduce un numero o escribe -1 para terminar'))
print(f'El numero es mas grande es: {numero_grande}')
````

````python
# imprimir todos los numeros del 1 a N 

numero = int(input('Introduce un numero'))

i = 1

while i <= numero:
    print(f'Soy el numero {i}')
    i = i + 1
print('fin del programa')
````

**Realizar un programa que calcule el promedio de calificaciones N** 

````python
calificacion = int(input('Introduce un numero'))

i = 1

while i <= calificacion:
    print(f'promedio {i}')
    i = i + i * i / i
print('promedio final')
````

Las sentencias breack y continue


````python
# breack sale del bucle inmediatamente e incondicionalmente termina la operacion del bucle, el programa empieza a ejecutar la instruccion mas cercana despues del cuerpo del bucle 
continue - se comporta como si el programa hubiera llegado repentinamente al final del cuerpo; el siguiente turno inicia y la exprecion de condicion se prueba de inmediato
````

````python
print("La instruccion break")
for i in range(1,6):
    if i == 3:
        break
    print('Dentro del bucle')
print('Fuera del bucle')
````

````python
# continue

print('\n la instruccion ccontinue')
for i in range(1,6):
    if i == 3:
        continue 
    print(f'Dentro del bucle {i}')
print('fuera del bucle')
````
